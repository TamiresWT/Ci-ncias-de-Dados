### Atividade 4: Crie uma coluna chamada "Days_Registered" que é a quantidade de dias desde que o cliente se registrou. Considere que a data de início é de 01/01/2015.

```python
#importando a biblioteca numpy
import numpy as np
```

```python
#importanto o módulo datetime, que fornece as classes para manipulação de datas e horas.
import datetime
```




*   Days_Registered: Quantidade de dias desde que o cliente se registrou
*   Dt_Customer: Data de cadastro do cliente



```python
#Passo 1: criando o dataframe date1 que estabelece a data inicial da contagem em 01/01/2015, através da função pd.to_datetime, e o método format
#para retornar as datas no formato dia/mês/ano com 4 dígitos
#Passo 2: criando o dataframe data2 somente com a coluna Dt_Customer do dataset df e insere o método format  para retornar as datas no formato
#dia/mês/ano com 4 dígitos
#Passo 3: criando a coluna Days_Registered no dataset df, que é a subtração do date2 do date1. O argumento np.timedelta64 é utilizado para
#definir a contagem em dias 'D' a partir de uma unidade de dia '1'


date1 = pd.to_datetime('01-01-2015', format = '%d-%m-%Y')
date2 = pd.to_datetime(df['Dt_Customer'], format =  '%d-%m-%Y')
df['Days_Registered'] = (date1 - date2) /np.timedelta64(1, 'D')

#Notas:
#O format usa algumas letras para representar as datas: %d = dia, %m = mês, %Y = ano com 4 dígitos,
#%y = ano com 2 dígitos, %H = horas, %M = minutos, %S = segundos e assim por diante.
#Outra possível solução, que é até mais simples, é adicionar o parâmetro 'dayfirst', que por padrão é False, e colocar como True:
#Ex. da bibliografia: df_setembro["Horário"]= pd.to_datetime(df_setembro["Horário"], dayfirst=True)
```

### Atividade 5: Crie um DataFrame apenas com as colunas `Recency`, `Days_Registered`, `MntTotal` e `Income` e use-o para fazer o agrupamento.

*   Recency: Número de dias desde a última compra do cliente
*   Days_Registered: Quantidade de dias desde que o cliente se registrou
*   "MntTotal": Valor gasto em frutas, carnes, doces, vinho e ouro nos últimos 2 anos
*   Income: Renda familiar anual do cliente

```python
# separando as colunas de interesse
df_filtro = df[['Recency','Days_Registered','MntTotal', 'Income']]
df_filtro.head()
```

### Atividade 6: Padronize os dados.

```python
#Padronização de dados - colocando na mesma escala#
# importando o scaler
from sklearn.preprocessing import StandardScaler

# instanciando o scaler
scaler = StandardScaler()

# criando uma cópia do DataFrame
df_scaled  = df_filtro.copy()

# aplicando a padronização
df_scaled[df_scaled.columns] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
# estatísticas descritivas do df_scaled
#função describe descreve as estatísticas
df_scaled.describe().T

#Notas:
#25% primeiro quartil
#50% mediana
#75% terceiro quartil
#valores máximos, minimos e a média
```

### Atividade 7: Faça um agrupamento hierárquico e plote o dendograma.

```python
# importando funções para plotar o dendograma

from scipy.cluster.hierarchy import linkage, dendrogram

#Notas:
#função linkage: calcula a distância entre os pontos
#função dendrogram:  cria o dendograma (representação gráfica)
```

```python
# gerando a matriz de distâncias com a função linkage

distancias = linkage(df_scaled[['Recency','Days_Registered','MntTotal','Income']],
                     method='ward',
                     metric='euclidean')

#Notas:
#objeto distancias é o resultado da função linkage - usar os dados a serem utilizados para calcular distâncias
#method:forma como o cálculo será processado
#metric: como cálcula (padrão: distância euclidean)
```

```python
#importanto a biblioteca seaborn
import seaborn as sns
sns.set(style='white', rc={'figure.figsize':(14,5)})

#contrução do dendograma 'dn'
dn = dendrogram(distancias,
                leaf_font_size=9,
                orientation='right')

#Notas:
#objeto dn recebe a função dendogram com as informações: objeto distancias calculado,
#parâmetro orientation como right para rotacionar o gráfico
#leaf_font_size controla o tamanho da fonte na folha (últimas linhas onde já aparecem os nomes dos estados)
```

### Atividade 8: Qual quantidade parece adequada para o número de grupos?

```python
# importando a função fcluster
from scipy.cluster.hierarchy import fcluster

# distância máxima
max_d = 38

#clusterização com base nas distâncias
# obtendo grupos levando em consideração a distância
grupos = fcluster(distancias, t=max_d, criterion='distance')
grupos
```

```python
#Clusterização por quantidade de clusters
# obtendo grupos levando em consideração a quantidade de grupos desejada
grupos = fcluster(distancias, t=3, criterion='maxclust')
grupos
```

```python
# adicionando a coluna grupos ao dataset df_scaled
df_scaled['grupos'] = grupos
df_scaled.head()
```

```python
#Plotando várias distribuições bivariadas de pares no conjunto de dados
sns.pairplot(df_scaled,
             hue= 'grupos', palette='tab10'
             );

```

### Para refletir: Como poderíamos usar uma variável categórica para realizar agrupamento?

Dica:  Pesquise sobre One Hot Encoding

Um método interessante é o uso do One Hot Encoding, que transforma variáveis categóricas em numéricas. Para o uso desse método, um primeiro passo pode ser reduzir ao máximo possível a quantidade de valores diferentes presentes na coluna de variável categórica. Em seguida, é preciso transformar cada valor presente na variável categórica numa nova coluna e transformá-la em coluna Dummy, ou seja, numa coluna numérica binária (0, 1). Isso é possível de ser realizado através da condição get_dummies da biblioteca pandas.
